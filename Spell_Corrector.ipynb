```python
from textblob import Word
from textblob import TextBlob
import pandas as pd
```

Load Dataset

```python
dataframe = pd.read_csv('mistakes.csv')
dataframe
```

Word and Sentence spell check

```python
def word_spell_check(word):
    word = Word(word)
    corrected_word = word.correct()
    return corrected_word

def sentence_spell_check(sentence):
    string = TextBlob(sentence)
    corrected_string = string.correct()
    return corrected_string
```

Spell Correcting

```python
correct_dataframe = pd.DataFrame({'correct':[]})
```

```python
for each_line in dataframe.iloc[:, 0]:
    length = len(each_line.split())
    spell_dataframe = pd.DataFrame({'correct':[str(sentence_spell_check(each_line)) if length > 1 else word_spell_check(each_line)]})
    correct_dataframe = correct_dataframe.append(spell_dataframe, ignore_index=True)
```

Result

```python
result = pd.concat([dataframe, correct_dataframe], axis=1).tail(100)
result
```
